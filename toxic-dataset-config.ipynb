```python
import numpy as np 
import pandas as pd 

input_path = '/kaggle/input'
output_path = '/kaggle/working/'
```

```python
df = pd.read_csv(input_path + '/toxicdb/Multi_labeled_toxic_comments.csv')
```

```python
df.head(5)
```

```python
# Select all columns except 'text'
toxic_identifying_cols = df.columns[1:]

# If all row values are '0' that means that instance is not toxic
non_toxic_df = df[(df[toxic_identifying_cols] == 0).all(axis=1)]

# Toxic instances would have at least one non-zero value
toxic_df = df[(df[toxic_identifying_cols] != 0).any(axis=1)]

# axis=1 implies row-wise operation, axis=0 implies column-wise operation
```

```python
print('Non Toxic Instances:\n')
print(non_toxic_df['text'].drop_duplicates())

print('Toxic Instances:\n')
print(toxic_df['text'].drop_duplicates())
```

```python
combined_df = pd.concat([
    hate_speech_df['text'], 
    toxic_df['text'], 
    non_toxic_df['text'], 
    # bangla_reviews_df,
    bd_shs_df
], ignore_index=True).drop_duplicates()

combined_df.to_csv(
    'combined_dataset.csv', 
    sep='\t',             # Tab-separated
    encoding='utf-8',     # Handle Bengali characters properly
    index=False
)
```

```python
from transformers import AutoTokenizer
from normalizer import normalize

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("csebuetnlp/banglabert")
```

```python
# Normalization
combined_df = combined_df.apply(normalize)

# Tokenization of each rows (text into subword tokens)
combined_df_tokens = combined_df.apply(lambda x: tokenizer.tokenize(x))

# Encode each rows (text into numerical IDs)
combined_df_input_ids = combined_df.apply(lambda x: tokenizer.encode(x, return_tensors="pt"))

processed_df = pd.DataFrame({
    'original_text': combined_df,
    'tokens': combined_df_tokens,
    'input_ids': combined_df_input_ids
})
```

```python
combined_df_tokens_lengths = combined_df_tokens.apply(len)

min_tokens = combined_df_tokens_lengths.min()
max_tokens = combined_df_tokens_lengths.max()

print("Min Token: ", min_tokens)
print("Max Token: ", max_tokens)

filtered_df = combined_df[combined_df_tokens_lengths > 6]
filtered_df.to_csv('filtered_dataset.csv', sep='\t', encoding='utf-8', index=False)

print(len(combined_df[(combined_df_tokens_lengths >= 101)]))
```
